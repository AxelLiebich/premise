```python
from premise import *
import bw2data, bw2calc
import pandas as pd
bw2data.projects.set_current("ei39")
```

```python
FU = [
    {a: 1} for a in
    bw2data.Database("test8")
    if a["name"].startswith("electricity production") and a["unit"] == "kilowatt hour"
]

print(len(FU))

list_meth = [
    ('IPCC 2021', 'climate change', 'GWP 100a, incl. H'),
    ('IPCC 2021', 'climate change', 'GWP 100a, incl. H and bio CO2'),
    ('selected LCI results', 'resource', 'land occupation'),
    ('EN15804', 'inventory indicators ISO21930', 'use of net fresh water'),
    ('Cumulative Energy Demand (CED)', 'total', 'energy content (HHV)'),
    ('USEtox', 'ecotoxicity', 'total')
]
import bw2calc
bw2data.calculation_setups['multiLCA'] = {'inv': FU, 'ia': list_meth}
myMultiLCA = bw2calc.MultiLCA('multiLCA')
df = pd.DataFrame(columns = [", ".join(i) for i in list_meth],
     data = myMultiLCA.results,
     index=[k["name"] + ", " + k["location"] for i in FU for k in i]
)
```

```python
import matplotlib.pyplot as plt

# Plot with labels only on the bottom row and shading for IQR
fig, axs = plt.subplots(3, 2, figsize=(12, 8))
axs = axs.flatten()  # Flatten to iterate easily

for i, column in enumerate(df.columns):
    sorted_df = df.sort_values(by=column)
    axs[i].plot(sorted_df.index, sorted_df[column], 'o', label='Values', zorder=1, markersize=0.5)
    mean_value = sorted_df[column].mean()
    median_value = sorted_df[column].median()
    axs[i].axhline(mean_value, color='r', linestyle='-', label=f'Mean: {mean_value:.4f}', zorder=2)
    axs[i].axhline(median_value, color='g', linestyle='--', label=f'Median: {median_value:.4f}', zorder=3)
    
    # Calculate and shade the IQR
    Q1 = sorted_df[column].quantile(0.25)
    Q3 = sorted_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    axs[i].fill_between(sorted_df.index, lower_bound, upper_bound, color='yellow', alpha=0.3, zorder=0, label='IQR Range')
    axs[i].set_xticklabels([])
    axs[i].set_title(column)
    axs[i].legend()

plt.tight_layout()
plt.show()

# Calculate the deviation of outliers from the mean instead of their absolute value
outliers_info_with_deviation = []

for column in df.columns:
    mean_value = df[column].mean()
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers_temp = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    for _, row in outliers_temp.iterrows():
        deviation_from_mean = row[column] / mean_value
        outliers_info_with_deviation.append({
            "Activity Name": row.name,
            "Impact Category": column,
            "Deviation from Mean": deviation_from_mean,
            "Value": row[column]
        })

outliers_df_with_deviation = pd.DataFrame(outliers_info_with_deviation)
```

```python
outliers_df_with_deviation.to_excel("outliers.xlsx")
```
